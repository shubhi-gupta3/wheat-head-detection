#### Importing Libraries

```
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```
